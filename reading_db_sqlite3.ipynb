```python
import sqlite3

# Conectando ao banco de dados SQLite (ou criando um novo banco de dados)
conn = sqlite3.connect('example.db')

# Criando um cursor para realizar operações no banco de dados
cursor = conn.cursor()

# Exemplo de criação de tabela (se ainda não existir)
cursor.execute('''
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER
)
''')

# Exemplo de inserção de dados
cursor.execute('INSERT INTO users (name, age) VALUES (?, ?)', ("Alice", 30))
cursor.execute('INSERT INTO users (name, age) VALUES (?, ?)', ("Bob", 24))
conn.commit()  # Salvando as mudanças

# Executando uma consulta
cursor.execute('SELECT * FROM users')

# Obtendo todos os resultados
rows = cursor.fetchall()

# Processando os resultados
for row in rows:
    print(f"ID: {row[0]}, Name: {row[1]}, Age: {row[2]}")

# Fechando a conexão
conn.close()
```
